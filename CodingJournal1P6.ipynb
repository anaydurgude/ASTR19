```python
import numpy as np
def sin(a):
    sinx = np.sin(a)
    return sinx
```

creates a variable sinx which is the sine of the input variable

```python
def cos(a):
    cosx = np.cos(a)
    return cosx
```

creates a variable cosx which is the cosine of the input variable

```python

x = np.linspace(0,2*np.pi,1000)
sinx=sin(x)
cosx=cos(x)
```

creates a variable x, which is a 1000 entries from 0 to 2pi, and then takes the sine and cosine of them

```python
for i in range(10):
    print(x[i],sinx[i],cosx[i])
```

prints the first 10 values of x, sinx, and cosx using a for loop
